```python
filename = '/home/roman/Downloads/Doner_Loren_Furi_(LP)_Litmir.net_bid224439_2a09f.txt'
try:
    with open(filename, 'r', encoding='utf-8') as f:
        content = f.read()
except UnicodeDecodeError:
    with open(filename, 'r', encoding='cp1251') as f:
        content = f.read()
except FileNotFoundError:
    content = 'Файл не найден!'
print(content)
```

```python
# Токенизация текста: разбивка на слова
import re
tokens = re.findall(r'\w+', content, flags=re.UNICODE)
print(tokens[:50])  # выводим первые 50 токенов
```

```python
# Токенизация текста с помощью nltk
import nltk

nltk.download('punkt_tab')
tokens_nltk = nltk.word_tokenize(content)
print(tokens_nltk[:50])  # первые 50 токенов
```

```python
# Лемматизация токенов с помощью pymorphy2
!pip install pymorphy2
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
lemmas = [morph.parse(token)[0].normal_form for token in tokens_nltk]
print(lemmas[:50])  # первые 50 лемм
```

```python
# Лемматизация токенов с помощью pymorphy2
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
lemmas = [morph.parse(token)[0].normal_form for token in tokens_nltk]
print(lemmas[:50])  # первые 50 лемм
```
