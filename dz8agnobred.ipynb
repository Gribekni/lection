с 1 агентом

```python
from dotenv import load_dotenv

from agno.agent import Agent
from ddgs import DDGS
import os

load_dotenv()

assert os.getenv("OPENAI_API_KEY"), "OPENAI_API_KEY not loaded"

def searchingtool(zapros: str) -> str:
    ddg = DDGS()
    lizt = list(ddg.text(zapros, max_results=5))
    otvet = ""
    for i, rez in enumerate(lizt, 1):
        otvet += f"{i}. {rez['title']}{rez['href']}{rez['body'][:200]}"
    return otvet

def agentsearching(zapros):
    agent = Agent(
        name="Searcher",
        tools=[searchingtool],
        instructions="Найди информацию в интернете по запросу и дай краткий ответ на русском.",
        markdown=False,
        model="openai:gpt-4o-mini",
    )
    otvet = agent.run(zapros)
    print(otvet.content)
    return otvet.content

agent_results = agentsearching("Что такое драм энд бейс?")
```

с Тимой

```python
from dotenv import load_dotenv
from agno.agent import Agent
from agno.team import Team
from ddgs import DDGS
import os

load_dotenv()

assert os.getenv("OPENAI_API_KEY"), "OPENAI_API_KEY not loaded"

def searchingtool(zapros: str) -> str:
    ddg = DDGS()
    lizt = list(ddg.text(zapros, max_results=5))
    otvet = ""
    for i, rez in enumerate(lizt, 1):
        otvet += f"{i}. {rez['title']}{rez['href']}{rez['body'][:200]}"
    return otvet

def searchwithteam(zapros):
    team = Team(
        name="SearchTeam",
        members=[
            Agent(
                name="Searcher",
                tools=[searchingtool],
                instructions="Найди информацию в интернете по запросу и дай краткий ответ на русском.",
                markdown=False,
                model="openai:gpt-4o-mini"
            ),
            Agent(
                name="Analyst",
                tools=[],
                instructions="Проанализируй информацию, полученную от исследователя, и дай краткий вывод на русском.",
                markdown=False,
                model="openai:gpt-4o-mini"
            )
        ],
    )
    otvet = team.run(zapros)
    print(otvet.content)
    return otvet.content

team_results = searchwithteam("Что такое драм энд бейс?")
```
