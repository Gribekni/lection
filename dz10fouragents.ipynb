```python
async def react(agent, message):
    result = await agent.arun(message)
    print(f"{agent.name}: {result.content}")
```

```python
async def SceneInShawshank():
    andysPhrase = "Порой надежда - это хорошая вещь, может быть, лучшая из вещей, а хорошие вещи никогда не умирают."
    print(f"Энди Дюфресн: {andysPhrase}")
    await react(Red, andysPhrase)
    await react(WardenNorton, andysPhrase)
    await react(CaptainHadley, andysPhrase)
    
    await asyncio.sleep(1)
    print()
    
    redsPhrase = "Надежда - это опасная вещь. Она может свести человека с ума."
    print(f"Ред: {redsPhrase}")
    await react(AndyDufresne, redsPhrase)
    await react(WardenNorton, redsPhrase)
    await react(CaptainHadley, redsPhrase)
    
    await asyncio.sleep(1)
    print()
    
    andysRoflPhrase = "Вот бы щас гидры поюзать"
    print(f"Энди Дюфресн: {andysRoflPhrase}")
    await react(Red, andysRoflPhrase)
    await react(WardenNorton, andysRoflPhrase)
    await react(CaptainHadley, andysRoflPhrase)
    
```
