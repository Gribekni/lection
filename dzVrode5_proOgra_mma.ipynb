```python
from typing import override

class MixinSpeakable:
    phrase: str = "urod"
    def speak(self) -> str:
        return self.phrase
    
class MixinAnimated:
    def isanimated(self) -> bool:
        return True

class MixinFunny:
    def make_laugh(self) -> str:
        return f"{self.name} makes us go hahaha"

class MixinIsLookingLikeADoofus:
    def is_looking_like_a_doofus(self) -> str:
        return f"{self.name} looks like a doofus"
    
class MixinActionable:
    def perform_action(self) -> str:
        return f"{self.name} doing something"

class MixinCollectible:
    def able_to_collect(self):
        return True

class MixinPoseable:
    def pose(self) -> str:
        return f"{self.name} striking a pose"
    
class MixinWearingACostume:
    def wear_costume(self, costume: str) -> str:
        return f"{self.name} is now wearing a costume"
    
class BaseCharacter(MixinSpeakable, MixinAnimated, MixinFunny):
    def __init__(self, name: str):
        self.name = name
    
class BaseHuman:
    def is_human(self) -> bool:
        return True
    def __init__(self, name: str):
        self.name = name

class BaseInActionCharacter(BaseCharacter):
    def perform_action(self) -> str:
        return f"{self.name} acts!"
    
class Shrek(BaseCharacter):
    name = "Shrek"
    phrase = "zATKNI HLEBALO I POISCHI GDE-NIBUD LESTNITSU!"

class Donkey(BaseCharacter):
    name = "Donkey"
    phrase = "Wow dragon, you're FIRE!"

class PussInBoots(BaseCharacter):
    name = "Puss in Boots"
    phrase = "Meow!"

class JackHorner(BaseCharacter):
    name = "Jack Horner"
    phrase = "Im fat lol."

#actions_menisk

class ShrekInAction(BaseInActionCharacter):
    def performs_action(self) -> str:
        return f"{self.name} is taking a mud bath!"
    
class DonkeyInAction(BaseInActionCharacter):
    def performs_action(self) -> str:
        return f"{self.name} is rizzing a dragon!"
    
class PussInBootsInAction(BaseInActionCharacter):
    def performs_action(self) -> str:
        return f"{self.name} is showing off his sword skills!"
    
class JackHornerInAction(BaseInActionCharacter):
    def performs_action(self) -> str:
        return f"{self.name} is being fat asf"
    
#figurki

class BaseFunkoPop(MixinCollectible, BaseCharacter):
    def justfigures(self) -> str:
        return "it is a " f"{self.name} Funko Pop collectible figure"
    
class ShrekFunkoPop(BaseFunkoPop):
    name = "Shrek"
    
class DonkeyFunkoPop(BaseFunkoPop):
    name = "Donkey"

class PussInBootsFunkoPop(BaseFunkoPop):
    name = "Puss in Boots"
    
class JackHornerFunkoPop(BaseFunkoPop):
    name = "Jack Horner"
    
#ryazhenieClowns

class BaseCosplayer(MixinPoseable, MixinWearingACostume, BaseHuman):
    def __init__(self, name: str):
        super().__init__(name)
        self.name = name

    def pose(self):
        text = f"{self.name} stands in a freaking COOL POSE!"
        if hasattr(self, "is_looking_like_a_doofus"):
            text += f" But {self.is_looking_like_a_doofus()}"
        print(text)

class ShrekCosplayer(BaseCosplayer):
    name = "Shrek"

class DonkeyCosplayer(MixinIsLookingLikeADoofus, BaseCosplayer):
    name = "Donkey"

class PussInBootsCosplayer(BaseCosplayer):
    name = "Puss in Boots"

class JackHornerCosplayer(BaseCosplayer):
    name = "Jack Horner"

#poehali

if __name__ == "__main__":
    
    shrek = Shrek(Shrek.name)
    donkey = Donkey(Donkey.name)
    puss_in_boots = PussInBoots(PussInBoots.name)
    jack_horner = JackHorner(JackHorner.name)

    shrek_in_action = ShrekInAction(Shrek.name)
    donkey_in_action = DonkeyInAction(Donkey.name)
    puss_in_boots_in_action = PussInBootsInAction(PussInBoots.name)
    jack_horner_in_action = JackHornerInAction(JackHorner.name)

    shrek_funko = ShrekFunkoPop(Shrek.name)
    donkey_funko = DonkeyFunkoPop(Donkey.name)
    puss_in_boots_funko = PussInBootsFunkoPop(PussInBoots.name)
    jack_horner_funko = JackHornerFunkoPop(JackHorner.name)

    shrek_cosplayer = ShrekCosplayer("Shrek")
    donkey_cosplayer = DonkeyCosplayer("Donkey")
    puss_in_boots_cosplayer = PussInBootsCosplayer("Puss in Boots")
    jack_horner_cosplayer = JackHornerCosplayer("Jack Horner")

    ssp = shrek.speak()
    dsp = donkey.speak()
    ppp = puss_in_boots.speak()
    jhp = jack_horner.speak()

    sia = shrek_in_action.performs_action()
    dia = donkey_in_action.performs_action()
    pia = puss_in_boots_in_action.performs_action()
    jha = jack_horner_in_action.performs_action()

    sff = shrek_funko.justfigures()
    dff = donkey_funko.justfigures()
    pff = puss_in_boots_funko.justfigures()
    jff = jack_horner_funko.justfigures()
    
    print("posing:")
    
    shrek_cosplayer.pose()
    donkey_cosplayer.pose()
    puss_in_boots_cosplayer.pose()
    jack_horner_cosplayer.pose()
    
    print("speaking:")
    
    print(ssp)
    print(dsp)
    print(ppp)
    print(jhp)
    
    print("actions:")
    
    print(sia)
    print(dia)
    print(pia)
    print(jha)
    
    print("funko pops:")
    
    print(sff)
    print(dff)
    print(pff)
    print(jff)

```
